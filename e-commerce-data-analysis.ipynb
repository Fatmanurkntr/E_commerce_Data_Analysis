<h2 style="color: #907ad6">2.3 Creating Missing Values (NaN) in the Dataset ‚ö†Ô∏è</h2>

<div style="background-color: #e7d8f7; padding: 10px; border-radius: 5px;">
In this section, we simulate missing values in our datasets to make our analysis more challenging and realistic. Real-world data is often incomplete or messy, and dealing with **missing values** is an essential part of data preprocessing. üßπ

### The Approach üí°
We define a function called `create_missing_values()` that introduces random missing values (NaN) into the datasets. Here's how it works:

- **Step 1**: We first calculate the total number of cells in the dataset using `np.prod(data.shape)`, which multiplies the number of rows by the number of columns.
  
- **Step 2**: We specify a `missing_rate`, which in this case is set to 5% (0.05). This rate determines what portion of the dataset will be turned into missing values.
  
- **Step 3**: For each cell to be made missing, we randomly select a row and a column index and set the corresponding cell to `NaN` (Not a Number), representing a missing value in our dataset.

### Applying the Function üöÄ
We apply this function to all three datasets: `sales_data`, `customer_details`, and `product_details`, introducing 5% missing values into each one. This helps us test how robust our analysis methods are in handling incomplete data.

- **`sales_data_missing`**: A new dataframe that contains the `sales_data` with 5% of its cells randomly turned into NaN values.
- **`customer_details_missing`**: Similarly, this dataframe holds the `customer_details` with some random missing values.
- **`product_details_missing`**: The `product_details` also undergoes the same treatment.

### Verifying Missing Values üîç
After creating the missing values, we use the `.isnull().sum()` method to count how many missing values are present in each column of the datasets. This gives us a clear idea of where the missing data is and how much of it exists.

This step allows us to handle a more realistic scenario where data is incomplete, which is often the case in real-world applications. Let's dive into data cleaning and handling these missing values next! üîß
</div>


```python
# Function: Creates random missing values with a specified rate
def create_missing_values(data, missing_rate=0.05):
    # Total number of cells in the dataset
    total_cells = np.prod(data.shape)  # np.product is replaced with np.prod
    
    # Number of cells to be left blank
    missing_cells = int(total_cells * missing_rate)
    
    # Randomly select row and column indices
    for _ in range(missing_cells):
        row = np.random.randint(0, data.shape[0])  # Random row selection
        col = np.random.randint(0, data.shape[1])  # Random column selection
        data.iat[row, col] = np.nan  # Set the selected cell to NaN

    return data

# Adding random missing values with a 5% rate to each dataset
sales_data_missing = create_missing_values(sales_data.copy(), missing_rate=0.05)
customer_details_missing = create_missing_values(customer_details.copy(), missing_rate=0.05)
product_details_missing = create_missing_values(product_details.copy(), missing_rate=0.05)

# Checking the datasets (shows the number of NaN values)
print("Sales Data with Missing Values:")
print(sales_data_missing.isnull().sum())

print("\nCustomer Details with Missing Values:")
print(customer_details_missing.isnull().sum())

print("\nProduct Details with Missing Values:")
print(product_details_missing.isnull().sum())
```

By removing these empty columns, we ensure that our datasets are leaner and contain only relevant data. This makes subsequent analyses more efficient and accurate. üöÄ

Now, let's check which columns were dropped and move forward with our cleaned datasets!
</div>


```python
# drop columns with 100% missing values and display the shape droped columns name
print("\n Sales Columns with 100% missing values: ", sales_data.columns[sales_data.isnull().all()])
sales_data.dropna(axis=1, how='all', inplace=True)

# drop columns with 100% missing values and display the shape droped columns name
print("\n Customer Details Columns with 100% missing values: ", customer_details.columns[customer_details.isnull().all()])
customer_details.dropna(axis=1, how='all', inplace=True)

# drop columns with 100% missing values and display the shape droped columns name
print("\n Product Details Columns with 100% missing values: ", product_details.columns[product_details.isnull().all()])
product_details.dropna(axis=1, how='all', inplace=True)
```

<h2 style="color: #907ad6"> 3.2 Checking for Missing Values in the Dataset üö®</h2>

<div style="background-color: #e7d8f7; padding: 10px; border-radius: 5px;">
In this step, we are examining missing values (NaN) in each dataset (Sales, Customer, Product) to understand the completeness of the data. The **`isnull()`** function checks for missing values in each cell, and **`sum()`** aggregates the count of missing values in each column. This allows us to see how many missing values are present in each column.

We use the following visualizations to make the missing values more apparent:
  
- **Bar Plot**: A bar chart that shows the number of missing values for each column in the dataset. 
- **Heatmap**: A heatmap that highlights missing values in a color-coded manner.
- **Missingno Visualization**: A visualization that provides a matrix view of missing values, which makes it easier to spot patterns of missingness across the dataset.

The visualizations are generated for all three datasets:
1. **Sales Data**: Missing values in sales-related information.
2. **Customer Details**: Missing values in customer-related data.
3. **Product Details**: Missing values in product-related data.
</div>


```python
import missingno as msno

# Visualizing missing values with a bar plot
def plot_missing_values_bar(data, title="Missing Values Bar Plot"):
    missing_data_count = data.isnull().sum()  # Calculate the number of missing values in each column
    missing_data_count = missing_data_count[missing_data_count > 0]  # Only columns with missing values
    
    if missing_data_count.empty:
        print(f"{title}: No missing values found.")  # Print a message if no missing values
        return
    
    plt.figure(figsize=(12, 6))
    missing_data_count.plot(kind='bar', color='tomato')  # Bar plot
    plt.title(title)
    plt.xlabel('Columns')
    plt.ylabel('Number of Missing Values')
    plt.show()

# Visualizing missing values with a heatmap
def plot_missing_values_heatmap(data, title="Missing Values Heatmap"):
    plt.figure(figsize=(12, 8))
    sns.heatmap(data.isnull(), cbar=False, cmap='viridis')  # Displays missing values in yellow/green tones
    plt.title(title)
    plt.show()

# Visualizing missing values with missingno
def plot_missing_values_missingno(data, title="Missing Values with Missingno"):
    plt.figure(figsize=(12, 8))
    msno.matrix(data)  # Visualizes missing values
    plt.title(title)
    plt.show()

# Visualizing missing values for each dataset
def plot_missing_values_for_datasets():
    # Sales Data
    plot_missing_values_bar(sales_data, "Sales Data Missing Values Bar Plot")
    plot_missing_values_heatmap(sales_data, "Sales Data Missing Values Heatmap")
    plot_missing_values_missingno(sales_data, "Sales Data Missing Values with Missingno")

    # Customer Details
    plot_missing_values_bar(customer_details, "Customer Details Missing Values Bar Plot")
    plot_missing_values_heatmap(customer_details, "Customer Details Missing Values Heatmap")
    plot_missing_values_missingno(customer_details, "Customer Details Missing Values with Missingno")

    # Product Details
    plot_missing_values_bar(product_details, "Product Details Missing Values Bar Plot")
    plot_missing_values_heatmap(product_details, "Product Details Missing Values Heatmap")
    plot_missing_values_missingno(product_details, "Product Details Missing Values with Missingno")

# Running the visualization function
plot_missing_values_for_datasets()
```

<h2 style="color: #907ad6">3.3 Filling Missing Data üõ†Ô∏è</h2>

<div style="background-color: #e7d8f7; padding: 10px; border-radius: 5px;">
Filling missing data is a critical step in data analysis and modeling. This process is essential to improve model accuracy, but the methods used may vary depending on the dataset. Below are the strategies we followed for each dataset, as well as explanations for why we did not fill the missing data in the Sales Data dataset.

<h4>1. Customer Details Dataset</h4>
In the Customer Details dataset, the following steps were taken to fill missing data:
<ul>
  <li><strong>Numerical Values</strong> (e.g., Age, Purchase Amount (USD)): Missing numerical values were filled with the mean of the existing values in the column. This approach is the most common method for filling missing values in numerical columns without disrupting the overall distribution of the data.</li>
  <li><strong>Categorical Values</strong> (e.g., Gender, Item Purchased, Subscription Status): For categorical columns, missing values were filled with the mode (most frequent value) of the column. Since calculating the mean for categorical data does not make sense, using the most frequent value is more logical.</li>
</ul>

<h4>2. Product Details Dataset</h4>
A similar approach was applied to the Product Details dataset:
<ul>
  <li><strong>Numerical Values</strong> (e.g., Selling Price, Shipping Weight): Missing values in numerical columns were filled with the mean. This allows for more meaningful evaluation of data such as product prices or weights.</li>
  <li><strong>Categorical Values</strong> (e.g., Product Name, Category): Missing values in categorical columns were filled with the mode. The most frequent value of categorical data can logically complete missing information.</li>
</ul>

<h4>3. Sales Data Dataset</h4>
The Sales Data dataset is the only dataset where missing data was not filled. There are several reasons for this:
<ul>
  <li><strong>Identity-Related Data</strong>: Columns such as user id and product id are unique identifiers for each entry, representing a person, product, or transaction. Filling missing values in these columns could disrupt the analysis process, as it is difficult to meaningfully impute these values. Especially since these columns are not used as key features for modeling, leaving them missing will not affect the analysis.</li>
  <li><strong>Timestamp Data</strong>: Timestamp data, such as time-related information, requires a specific order and sequence. Filling missing values in timestamp columns could disrupt the time sequence, making it difficult to conduct accurate analysis. In time-series data, filling missing values is typically done in a way that does not affect the chronological order, so it was preferred to leave these columns with missing data.</li>
</ul>

<h4><strong>Why Wasn't the Sales Data Filled?</strong></h4>
<ul>
  <li><strong>Identity Data</strong>: user id, product id, and similar data are unique identifiers for each transaction. Missing values in these columns could lead to errors in analysis or modeling because it is challenging to impute them meaningfully. Filling them with incorrect values could mislead the model‚Äôs learning. Therefore, instead of filling missing data in these columns, it was preferred to ignore these missing values in the analysis process.</li>
  <li><strong>Timestamp Data</strong>: Time-related data is associated with a specific order or time period. When the timestamp is missing, filling this data with methods like the mean or mode is not meaningful because it would disrupt the time sequence, hindering the accuracy of the analysis. Therefore, missing timestamp data was left as is to preserve the chronological order.</li>
</ul>
</div>


```python
# For customer_details, fill missing numeric values with the mean and categorical values with the mode
for column in customer_details.columns:
    if customer_details[column].dtype == 'object':
        customer_details[column].fillna(customer_details[column].mode()[0], inplace=True)
    else:
        customer_details[column].fillna(customer_details[column].mean(), inplace=True)

# For product_details, assuming missing values in 'List Price' and other numeric fields, fill with mean
# and categorical fields with mode
for column in product_details.columns:
    if product_details[column].dtype == 'object':
        product_details[column].fillna(product_details[column].mode()[0], inplace=True)
    else:
        product_details[column].fillna(product_details[column].mean(), inplace=True)

# Summarizing key statistics for numerical and categorical variables
sales_summary = sales_data.describe(include=[np.number])
customer_summary = customer_details.describe(include=[np.number])

sales_cat_summary = sales_data.describe(include=[object])
customer_cat_summary = customer_details.describe(include=[object])
product_cat_summary = product_details.describe(include=[object])

# Displaying summaries
print("Sales Data Summary (Numerical):")
print(sales_summary)
print("\nSales Data Summary (Categorical):")
print(sales_cat_summary)

print("\nCustomer Details Summary (Numerical):")
print(customer_summary)
print("\nCustomer Details Summary (Categorical):")
print(customer_cat_summary)
```

<h3 style="color: #907ad6">First Merge - sales_data and customer_details:</h3>

<div style="background-color: #e7d8f7; padding: 10px; border-radius: 5px;">
    <p>In this step, we are merging the <strong>sales_data</strong> and <strong>customer_details</strong> datasets. The merge operation is based on matching the 'user id' column in <strong>sales_data</strong> with the 'Customer ID' column in <strong>customer_details</strong>. The <strong>how='inner'</strong> parameter ensures that only the rows with matching 'user id' and 'Customer ID' values in both datasets are merged. This means that only customers present in both datasets, along with their sales data, will be merged. As a result, a new dataframe called <strong>sales_customer</strong> is created, which contains both customer and sales information.</p>
</div>


```python
sales_customer = pd.merge(sales_data, customer_details, left_on='user id', right_on='Customer ID', how='inner')
```

<h3 style="color: #907ad6">Second Merge - sales_customer and product_details:</h3>

<div style="background-color: #e7d8f7; padding: 10px; border-radius: 5px;">
    <p>In this step, we are merging the <strong>sales_customer</strong> and <strong>product_details</strong> datasets. The merge operation is performed based on matching the 'product id' column in <strong>sales_customer</strong> with the 'Uniqe Id' column in <strong>product_details</strong>. Again, using the <strong>how='inner'</strong> parameter ensures that only the rows with matching 'product id' and 'Uniqe Id' values in both datasets are merged.</p
    <p>As a result, a new dataframe called <strong>sales_customer_product</strong> is created. This dataframe contains sales data, customer information, and product details, allowing for a comprehensive view of the sales transactions.</p>
</div>


```python
sales_customer_product = pd.merge(sales_customer, product_details, left_on='product id', right_on='Uniqe Id', how='inner')
```

<h3 style="color: #907ad6">Processing of Numerical Features üî¢</h3>
<div style="background-color:#e7d8f7;; padding: 10px; border-radius: 5px;">
    <p>Some numerical columns (e.g., Age, Purchase Amount (USD), Review Rating, Previous Purchases, Hour) can be directly included in the model. However, columns like <strong>Purchase Amount (USD)</strong> are features that need to be scaled.</p>
    <h4>Steps:</h4>

```python
from sklearn.preprocessing import StandardScaler

# Create a copy of the original dataset
sales_customer_product_copy = sales_customer_product.copy()

# Select numeric columns
numeric_columns = sales_customer_product_copy.select_dtypes(include=['number']).columns

# Scaling process
scaler = StandardScaler()
sales_customer_product_copy[numeric_columns] = scaler.fit_transform(sales_customer_product_copy[numeric_columns])

# Now the numeric data in 'sales_customer_product_copy' has been scaled
```

<h3 style="color: #907ad6">Processing of Categorical Features üîÑ</h3>
<div style="background-color:#e7d8f7;; padding: 10px; border-radius: 5px;">
    <p>Categorical data must be converted into numerical data so that the model can process it. One common approach is to use <strong>One-Hot Encoding.</strong></p>
    <h4>Steps:</h4>
    <ul>
        <li><strong>Converting Categorical Features using One-Hot Encoding:</strong> Categorical columns are converted into numerical data using <code>pd.get_dummies()</code>.</li>
    </ul>
</div>


```python
from sklearn.preprocessing import StandardScaler
import pandas as pd

# Create a copy of the original dataset
sales_customer_product_copy = sales_customer_product.copy()

# Select numeric columns and apply scaling (from the previous step)
numeric_columns = sales_customer_product_copy.select_dtypes(include=['number']).columns
scaler = StandardScaler()
sales_customer_product_copy[numeric_columns] = scaler.fit_transform(sales_customer_product_copy[numeric_columns])

# Select categorical columns
categorical_columns = sales_customer_product_copy.select_dtypes(include=['object']).columns

# Apply One-Hot Encoding
sales_customer_product_copy = pd.get_dummies(sales_customer_product_copy, columns=categorical_columns)

# Now 'sales_customer_product_copy' has both standardization and One-Hot Encoding applied
```

<h3 style="color: #907ad6">Feature Construction üîß</h3>
<div style="background-color:#e7d8f7;; padding: 10px; border-radius: 5px;">
    <p>You can create more meaningful features in your dataset. For example:</p>
    <ul>
        <li><strong>Creating Age Groups:</strong> Dividing the <code>Age</code> column into categories can help the model learn better.</li>
        <li><strong>Purchase Frequency:</strong> A new feature representing the customer's shopping habits can be created using the <code>Purchase Amount (USD)</code> and <code>Previous Purchases</code> columns.</li>
    </ul>
</div>


```python
# Age Groups
bins = [0, 18, 30, 40, 50, 60, 100]
labels = ['0-18', '19-30', '31-40', '41-50', '51-60', '60+']
sales_customer_product_copy['Age Group'] = pd.cut(sales_customer_product_copy['Age'], bins=bins, labels=labels)

# Purchase Frequency (e.g., based on previous purchase count)
sales_customer_product_copy['Purchase Frequency'] = sales_customer_product_copy['Previous Purchases'] / sales_customer_product_copy['Purchase Amount (USD)']

# Now the dataset includes new features like 'Age Group' and 'Purchase Frequency'
```

<h3 style="color: #907ad6"> Binning (Feature Grouping) üìä</h3>
<div style="background-color:#e7d8f7; ; padding: 10px; border-radius: 5px;">
    <p>Converting numerical data into categorical data can be useful in some cases. For example, by dividing columns like <code>Review Rating</code> into specific ranges, you can make it easier for the model to understand.</p>
</div>


```python
# Review Rating Binning
review_bins = [0, 2, 4, 5]
review_labels = ['Low', 'Medium', 'High']
sales_customer_product_copy['Review Rating Group'] = pd.cut(sales_customer_product_copy['Review Rating'], bins=review_bins, labels=review_labels)

# Now there is a new feature called 'Review Rating Group'
```

<h2 style="color: #907ad6">5.2 Feature Selection üß© </h2>
<div style="background-color:#e7d8f7;; padding: 10px; border-radius: 5px;">
    <p><strong>Objective:</strong></p>
    <ul>
        <li><strong>Reduce the Dataset:</strong> Train the model with more meaningful features.</li>
        <li><strong>Avoid Overfitting:</strong> Prevent the model from overfitting by removing unnecessary or highly correlated features.</li>
        <li><strong>Speed and Efficiency:</strong> Reducing the number of features shortens training time and increases model efficiency.</li>
    </ul>
    <p><strong>Methods:</strong></p>
    <ul>
        <li><strong>Correlation-Based Selection:</strong> Examine correlations between features and remove those that have high correlation.</li>
        <li><strong>Feature Importance Scores:</strong> Model-based selections (e.g., feature importance scores from models like Random Forest, XGBoost).</li>
        <li><strong>Recursive Feature Elimination (RFE):</strong> A method that removes features one by one.</li>
        <li><strong>Removing Redundant Features:</strong> Columns that provide the same or very similar information can be removed.</li>
    </ul>
</div>


<h3 style="color: #907ad6">Correlation-Based Selection üîó</h3>
<div style="background-color:#e7d8f7; ; padding: 10px; border-radius: 5px;">
    <p>By detecting high correlations between features, we can remove those that are very similar. This helps reduce the complexity of the model.</p>
    <p><strong>Steps:</strong></p>
    <ul>
        <li>Examine columns with high correlation and set a threshold (e.g., remove features with a correlation of 0.8 or higher).</li>
    </ul>
</div>


```python
# Create a copy of the data
sales_customer_product_corr = sales_customer_product.copy()

# Select numeric columns
numeric_columns = sales_customer_product_corr.select_dtypes(include=['number']).columns

# Calculate the correlation matrix
corr_matrix = sales_customer_product_corr[numeric_columns].corr()

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()

# Select features with high correlation (excluding its own column)
threshold = 0.8  # Correlation threshold
high_corr_features = [col for col in corr_matrix.columns if any(corr_matrix[col] > threshold) and col != corr_matrix.columns[corr_matrix[col].argmax()]]

# Remove features with high correlation from the dataset
sales_customer_product_corr = sales_customer_product_corr.drop(columns=high_corr_features)

# Print the removed high correlation features
print(f"Removed high correlation features: {high_corr_features}")
```

<h3 style="color: #907ad6">Feature Selection Based on Feature Importance ‚ö°</h3>
<div style="background-color: #e7d8f7; padding: 10px; border-radius: 5px;">
    <p><strong>Explanation:</strong></p>
    <ul>
        <li><strong>Feature Importance:</strong> Feature importance refers to the contribution of each feature to the prediction of the target variable. After training a model, such as Random Forest or XGBoost, we can calculate the importance of each feature based on how much it helps in reducing prediction error.</li>
        <li><strong>Reduce Model Complexity:</strong> Not all features contribute equally to the model's performance. Features with low importance can be removed to reduce model complexity, making it simpler and more interpretable. A simpler model also helps to avoid overfitting, where the model learns too much from noise or irrelevant features.</li>
        <li><strong>Efficiency:</strong> Removing less important features reduces the number of variables that the model has to process, which in turn speeds up training time. It also reduces the computational resources required for model training, making the model more efficient overall.</li>
    </ul>
    <p><strong>Why it's done:</strong></p>
    <ul>
        <li>By identifying and removing less important features, we ensure that the model focuses only on the most relevant data, improving both its predictive power and interpretability.</li>
        <li>This process helps improve the model‚Äôs accuracy by preventing it from relying on unnecessary or noisy features that could hinder its ability to generalize well to unseen data.</li>
        <li>Optimizing model performance and reducing overfitting are key goals in machine learning, and feature selection based on importance helps achieve both of these goals effectively.</li>
    </ul>
    <p><strong>Steps to Implement:</strong></p>
    <ul>
        <li><strong>Train a Model:</strong> Use models like Random Forest or XGBoost, which can calculate feature importance based on their internal structure.</li>
        <li><strong>Calculate Feature Importance:</strong> Once the model is trained, obtain the feature importance scores, which rank features according to how much they contribute to the prediction.</li>
        <li><strong>Remove Low Importance Features:</strong> Based on the calculated importance scores, remove the features with low importance, keeping only those that significantly contribute to the model‚Äôs performance.</li>
    </ul>
</div>
</ul>
</div>
